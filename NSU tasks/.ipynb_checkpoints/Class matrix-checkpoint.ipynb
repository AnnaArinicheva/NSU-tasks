```python
#include <iostream>

using namespace std;

class matrix {
private: int** table;
public: int dim;
public:
    matrix();
    ~matrix();
    matrix(int);
    matrix(int, int*);
    matrix(const matrix& b);
    void Output_matrix();
    void Input_matrix(int);
    matrix operator+ (const matrix&) const;
    matrix operator- (const matrix&) const;
    matrix operator* (const matrix&) const;
    bool operator== (const matrix&) const;
    bool operator!= (const matrix&) const;
    matrix operator! () const;
    matrix operator() (const int, const int) const;
    matrix& operator= (const matrix&);
    int* operator[] (int) const;


    class column_of_matrix {
    private:
        int** column;
    public:
        int dimention;
        column_of_matrix(int index, matrix& a);
        ~column_of_matrix();
        column_of_matrix(const column_of_matrix& b);
        column_of_matrix& operator= (const column_of_matrix&);
        int& operator[](int index);
    };

    matrix::column_of_matrix operator() (int);

};



matrix::matrix() {
    dim = 0;
    table = NULL;
};

matrix::~matrix() {
    for (int i = 0; i < dim; i++) {
        delete[] table[i];
    }
    delete[] table;
    dim = 0;
};

matrix& matrix::operator= (const matrix& b) {
    for (int i = 0; i < dim; i++) {
        delete[] table[i];
    }
    delete[] table;
    dim = b.dim;
    table = new int* [b.dim];
    for (int i = 0; i < b.dim; i++) {
        table[i] = new int[b.dim];
        for (int j = 0; j < b.dim; j++) {
            table[i][j] = b.table[i][j];
        }
    }
    return *this;
};

matrix::matrix(const matrix& b) {
    dim = b.dim;
    table = new int* [b.dim];
    for (int i = 0; i < b.dim; i++) {
        table[i] = new int[b.dim];
        for (int j = 0; j < b.dim; j++) {
            table[i][j] = b.table[i][j];
        }
    }
};

matrix::matrix(int n) {
    dim = n;
    table = new int* [n];
    for (int i = 0; i < n; i++) {
        table[i] = new int[n];
        for (int j = 0; j < n; j++) {
            if (i == j)
                table[i][j] = 1;
            else
                table[i][j] = 0;
        }
    }
};


matrix::matrix(int n, int* set) {
    dim = n;
    table = new int* [n];
    for (int i = 0; i < n; i++) {
        table[i] = new int[n];
        for (int j = 0; j < n; j++) {
            if (i == j)
                table[i][j] = set[i];
            else
                table[i][j] = 0;
        }
    }
};

void matrix::Output_matrix() {
    for (int i = 0; i < dim; i++) {
        for (int j = 0; j < dim; j++) {
            cout << table[i][j] << ' ';
        }
        cout << endl;
    }
};

void matrix::Input_matrix(int n) {
    dim = n;
    table = new int* [n];
    for (int i = 0; i < n; i++) {
        table[i] = new int[n];
        for (int j = 0; j < dim; j++) {
            cin >> table[i][j];
        }
    }
};

matrix matrix::operator+ (const matrix& b) const {
    if (dim != b.dim) {
        cout << "Matrices have different sizes" << endl;
        matrix c;
        return c;
    }
    else {
        matrix c(dim);
        for (int i = 0; i < dim; i++) {
            for (int j = 0; j < dim; j++) {
                c.table[i][j] = table[i][j] + b.table[i][j];
            }
        }
        return c;
    }
};

matrix matrix::operator- (const matrix& b) const {
    if (dim != b.dim) {
        cout << "Matrices have different sizes" << endl;
        matrix c;
        return c;
    }
    else {
        matrix c(dim);
        for (int i = 0; i < dim; i++) {
            for (int j = 0; j < dim; j++) {
                c.table[i][j] = table[i][j] - b.table[i][j];
            }
        }
        return c;
    }
};

matrix matrix::operator* (const matrix& b) const {
    if (dim != b.dim) {
        cout << "Matrices have different sizes" << endl;
        matrix c;
        return c;
    }
    else {
        matrix c(dim);
        for (int i = 0; i < dim; i++) {
            for (int j = 0; j < dim; j++) {
                c.table[i][j] = 0;
                for (int k = 0; k < dim; k++) {
                    c.table[i][j] = c.table[i][j] + (table[i][k] * b.table[k][j]);
                }
            }
        }
        return c;
    }
};

bool matrix::operator== (const matrix& b) const {
    if (dim != b.dim) {
        return false;
    }
    else {
        for (int i = 0; i < dim; i++)
            for (int j = 0; j < dim; j++)
                if (table[i][j] != b.table[i][j])
                    return false;
    }
    return true;
};

bool matrix::operator!= (const matrix& b) const {
    if (dim != b.dim) {
        return true;
    }
    else {
        for (int i = 0; i < dim; i++)
            for (int j = 0; j < dim; j++)
                if (table[i][j] != b.table[i][j])
                    return true;
    }
    return false;
};

matrix matrix::operator! () const {
    matrix c(dim);
    for (int i = 0; i < dim; i++) {
        for (int j = 0; j < dim; j++) {
            c.table[i][j] = table[j][i];
        }
    }
    return c;
};

matrix matrix::operator() (const int a, const int b) const {
    matrix c(dim-1);
    int i = 0, j = 0, i_of_this = 0, j_of_this = 0;
    for (i_of_this = 0; i_of_this < dim; i_of_this++) {
        if (i_of_this != (a - 1)) {
            for (j_of_this = 0; j_of_this < dim; j_of_this++) {
                if (j_of_this != (b - 1)) {
                    c.table[i][j] = table[i_of_this][j_of_this];
                    j = (j + 1) % 3;
                }
            }
            i = (i + 1) % 3;
        }
    }
    return c;
};


int* matrix::operator[] (int a) const {
    return table[a - 1];
};

int& matrix::column_of_matrix::operator[] (int a) {
    return *(column[a - 1]);
};

matrix::column_of_matrix matrix::operator()(int b) {
    column_of_matrix col(b, *this);
    return col;
};







matrix::column_of_matrix::column_of_matrix(int index, matrix& a) {
    dimention = a.dim;
    column = new int* [a.dim];
    for (int i = 0; i < a.dim; i++)
        column[i] = &a.table[i][index - 1];
};

matrix::column_of_matrix::~column_of_matrix() {
    for (int i = 0; i < dimention; i++) {
        delete[] column[i];
    }
    delete[] column;
    dimention = 0;
};

matrix::column_of_matrix::column_of_matrix(const column_of_matrix& b) {
    dimention = b.dimention;
    column = new int* [b.dimention];
    for (int i = 0; i < dimention; i++) {
        column[i] = b.column[i];
    }
};

matrix::column_of_matrix& matrix::column_of_matrix::operator= (const column_of_matrix& b) {
    for (int i = 0; i < dimention; i++) {
        delete[] column[i];
    }
    delete[] column;
    dimention = b.dimention;
    column = new int* [b.dimention];
    for (int i = 0; i < dimention; i++) {
        column[i] = b.column[i];
    }
    return *this;
};








int main() {
    int n, k;
    cin >> n;
    cin >> k;
    matrix A, B, C, D, ANS;
    A.Input_matrix(n);
    B.Input_matrix(n);
    C.Input_matrix(n);
    D.Input_matrix(n);
    int* diag = new int[n];
    for (int i = 0; i < n; i++) {
        diag[i] = k;
    }
    matrix K(n, diag);
    ANS = (A + B * (!(C)) + K) * (!(D));
    cout << endl;
    ANS.Output_matrix();


    /*
    matrix a;
    a.Input_matrix(4);
    int *b = new int(4);
    b = a[2];
    cout << endl;
    for (int i = 0; i < 4; i++){
      cout << b[i] << ' ';
    }

    cout << endl;

    int *c = new int(4);
    c = a(3);
    for (int i = 0; i < 4; i++){
      cout << c[i] << ' ';
    }
    cout << endl;
    cout << a(2)[3] << endl;

    a(2)[2] = 12;
    cout << endl;
    a.Output_matrix();
    */

    return 0;
}
```
