```python
#include <iostream>
#include <string>

using namespace std;

//класс матриц из первого задания
class matrix {
public: int** table;
public: int dim;
public:
	matrix();
	~matrix();
	matrix(int);
	matrix(int, int*);
	matrix(const matrix& b);
	void Output_matrix();
	void Input_matrix(int);
	matrix operator+ (const matrix&) const;
	matrix operator- (const matrix&) const;
	bool operator== (const matrix&) const;
	bool operator!= (const matrix&) const;
	matrix& operator= (const matrix&);
};



matrix::matrix() {
	dim = 0;
	table = NULL;
};

matrix::~matrix() {
	for (int i = 0; i < dim; i++) {
		delete[] table[i];
	}
	delete[] table;
	dim = 0;
};

matrix& matrix::operator= (const matrix& b) {
	for (int i = 0; i < dim; i++) {
		delete[] table[i];
	}
	delete[] table;
	dim = b.dim;
	table = new int* [b.dim];
	for (int i = 0; i < b.dim; i++) {
		table[i] = new int[b.dim];
		for (int j = 0; j < b.dim; j++) {
			table[i][j] = b.table[i][j];
		}
	}
	return *this;
};

matrix::matrix(const matrix& b) {
	dim = b.dim;
	table = new int* [b.dim];
	for (int i = 0; i < b.dim; i++) {
		table[i] = new int[b.dim];
		for (int j = 0; j < b.dim; j++) {
			table[i][j] = b.table[i][j];
		}
	}
};

matrix::matrix(int n) {
	dim = n;
	table = new int* [n];
	for (int i = 0; i < n; i++) {
		table[i] = new int[n];
		for (int j = 0; j < n; j++) {
			if (i == j)
				table[i][j] = 1;
			else
				table[i][j] = 0;
		}
	}
};


matrix::matrix(int n, int* set) {
	dim = n;
	table = new int* [n];
	for (int i = 0; i < n; i++) {
		table[i] = new int[n];
		for (int j = 0; j < n; j++) {
			if (i == j)
				table[i][j] = set[i];
			else
				table[i][j] = 0;
		}
	}
};

void matrix::Output_matrix() {
	for (int i = 0; i < dim; i++) {
		for (int j = 0; j < dim; j++) {
			cout << table[i][j] << ' ';
		}
		cout << endl;
	}
};

void matrix::Input_matrix(int n) {
	dim = n;
	table = new int* [n];
	for (int i = 0; i < n; i++) {
		table[i] = new int[n];
		for (int j = 0; j < dim; j++) {
			cin >> table[i][j];
		}
	}
};

matrix matrix::operator+ (const matrix& b) const {
	if (dim != b.dim) {
		cout << "Matrices have different sizes" << endl;
		matrix c;
		return c;
	}
	else {
		matrix c(dim);
		for (int i = 0; i < dim; i++) {
			for (int j = 0; j < dim; j++) {
				c.table[i][j] = table[i][j] + b.table[i][j];
			}
		}
		return c;
	}
};

matrix matrix::operator- (const matrix& b) const {
	if (dim != b.dim) {
		cout << "Matrices have different sizes" << endl;
		matrix c;
		return c;
	}
	else {
		matrix c(dim);
		for (int i = 0; i < dim; i++) {
			for (int j = 0; j < dim; j++) {
				c.table[i][j] = table[i][j] - b.table[i][j];
			}
		}
		return c;
	}
};


bool matrix::operator== (const matrix& b) const {
	if (dim != b.dim) {
		return false;
	}
	else {
		for (int i = 0; i < dim; i++)
			for (int j = 0; j < dim; j++)
				if (table[i][j] != b.table[i][j])
					return false;
	}
	return true;
};

bool matrix::operator!= (const matrix& b) const {
	if (dim != b.dim) {
		return true;
	}
	else {
		for (int i = 0; i < dim; i++)
			for (int j = 0; j < dim; j++)
				if (table[i][j] != b.table[i][j])
					return true;
	}
	return false;
};



//специальное хэширование для матриц
template<>
struct hash<matrix> {
	size_t operator()(matrix const& a) const noexcept {
		int sum = 0;
		for (int i = 0; i < a.dim; i++) {
			for (int j = 0; j < a.dim; j++) {
				sum += a.table[i][j];
			}
		}
		return sum;
	}
};



//реализация третьего задания:


//класс элементов - пар
template <typename K, typename V>
class Elements {
public:
	K _key;
	V _value;
	bool flag;
	Elements(K key, V value);
	~Elements() = default;
};

//класс таблицы
template <typename K, typename V>
class Hash_Map {
protected:
	Elements<K, V>** table_elements;
	int size_table = 0;
	int count_elements = 0;
	int occupancy = 75;
public:
	Hash_Map(int size);
	~Hash_Map();
	void insert(K key, V value);
	void collision(K key, V value, Elements<K, V>* element);
	V search(K key) const;
	void print_table() const;
	void delete_element(K key);
	void restructuring();
	void unique();


	//служебный класс итератор
	class Iterator {
	public:
		Elements<K, V>* _element;
	public:
		Hash_Map<K, V>* hashmap;
		int _position = 0;
		Iterator(Elements<K, V>* element1, Hash_Map* map, int position);
		Elements<K, V> operator*() const;
		Iterator operator++();
		bool operator != (const Iterator element2) const;
		bool operator == (const Iterator element2) const;
		bool get_flag() const;
		K get_key() const;
		V get_value() const;

	};

	Iterator begin() const;
	Iterator end() const;
};


//конструктор элементов - пар
template <typename K, typename V>
Elements<K, V>::Elements(K key, V value) {
	_key = key;
	_value = value;
	bool flag = true;
}


//конструктор таблицы
template <typename K, typename V>
Hash_Map<K, V>::Hash_Map(int size) {
	size_table = size;
	count_elements = 0;
	table_elements = new Elements<K, V>*[size];
	for (int i = 0; i < size; i++) {
		table_elements[i] = NULL;
	}
}

//деструктор таблицы
template <typename K, typename V>
Hash_Map<K, V>::~Hash_Map() {
	for (int i = 0; i < size_table; i++) {
		if (table_elements[i] != NULL)
			delete table_elements[i];
	}
	delete[] table_elements;
	size_table = 0;
	count_elements = 0;
}



//конструктор итератора
template <typename K, typename V>
Hash_Map<K, V>::Iterator::Iterator(Elements<K, V>* element, Hash_Map* map, int position) {
	_element = element;
	hashmap = map;
	_position = position;
};

//разыменовывание
template <typename K, typename V>
Elements<K, V> Hash_Map<K, V>::Iterator::operator*() const {
	return &_element;
}

//++
template <typename K, typename V>
typename Hash_Map<K, V>::Iterator Hash_Map<K, V>::Iterator:: operator++() {
	for (int i = 0; i < hashmap->size_table; i++) {
		if (i == _position && i != hashmap->size_table - 1) {
			this->_element = hashmap->table_elements[i + 1];
			this->_position++;
			return *this;
		}
		else if (i == _position && i == hashmap->size_table - 1) {
			this->_element = NULL;
			this->_position++;
			return *this;
		}
	}
}

//!=
template <typename K, typename V>
bool Hash_Map<K, V>::Iterator::operator != (const Iterator element2) const{
	if (_position == element2._position) {
		return false;
	}
	else {
		return true;
	}
}

//==
template <typename K, typename V>
bool Hash_Map<K, V>::Iterator::operator == (const Iterator element2) const{
	if (_position == element2._position) {
		return true;
	}
	else {
		return false;
	}
}


//begin
template <typename K, typename V>
typename Hash_Map<K, V>::Iterator Hash_Map<K, V>::begin() const {
	Hash_Map<K, V>::Iterator iter(table_elements[0], this, 0);
	return iter;
}

//end
template <typename K, typename V>
typename Hash_Map<K, V>::Iterator Hash_Map<K, V>::end() const {
	Hash_Map<K, V>::Iterator iter(NULL, this, size_table);
	return iter;
}

//get
template <typename K, typename V>
bool Hash_Map<K, V>::Iterator::get_flag() const {
	return this->_element->flag;
}
template <typename K, typename V>
K Hash_Map<K, V>::Iterator::get_key() const {
	return this->_element->_key;
}
template <typename K, typename V>
V Hash_Map<K, V>::Iterator::get_value() const {
	return this->_element->_value;
}




//вывод таблицы на экран
template <typename K, typename V>
void Hash_Map<K, V>::print_table() const {
	cout << endl;
	cout << endl;
	cout << "PRINT TABLE:" << endl;
	
	Hash_Map<K, V>::Iterator iter(table_elements[0], this, 0);
	for (iter = this->begin(); iter != this->end(); ++iter) {
		if (iter._element != NULL && iter.get_flag() != false) {
			cout << iter.get_key();
			cout << " ";
			cout << iter.get_value() << endl;
		}
		else {
			cout << 0 << endl;
		}
	}
	
}

//Метод открытой адресации для разрешения коллизий
template <typename K, typename V>
void Hash_Map<K, V>::collision(K key, V value, Elements<K, V>* element) {

	size_t index = hash<K>()(key) % size_table;
	for (int i = 0; i < size_table; i++) {
		int cur_index = (index + i) % size_table;
		Elements<K, V>* cur_element = table_elements[cur_index];
		if (cur_element == NULL) {

			table_elements[cur_index] = element;
			table_elements[cur_index]->flag = true;
			count_elements++;
			if (count_elements / size_table * 100 >= occupancy) {
				this->restructuring();
			}
			break;
		}
	}

}

//поиск элемента по ключу
template <typename K, typename V>
V Hash_Map<K, V>::search(K key) const {
	size_t index = hash<K>()(key) % size_table;
	for (int i = 0; i < size_table; i++) {
		Elements<K, V>* cur_element = table_elements[(index + i) % size_table];
		if (cur_element != NULL) {
			if (cur_element->_key == key && cur_element->flag != false) {
				return cur_element->_value;
			}
		}
	}
	return 0;
}

//удаление элемента из таблицы
template <typename K, typename V>
void Hash_Map<K, V>::delete_element(K key) {
	size_t index = hash<K>()(key) % size_table;

	for (int i = 0; i < size_table; i++) {
		int cur_index = (index + i) % size_table;
		Elements<K, V>* cur_element = table_elements[cur_index];
		if (cur_element != NULL && cur_element->flag != false) {
			if (cur_element->_key == key) {
				table_elements[cur_index]->flag = false;
				count_elements--;
			}
		}
	}
}

//перестройка таблицы для поддержания разреженности
template <typename K, typename V>
void Hash_Map<K, V>::restructuring() {
	size_table *= 2;
	Elements<K, V>** table = new Elements<K, V>*[size_table];
	for (int i = 0; i < size_table / 2; i++) {
		if (table_elements[i] != NULL) {
			Elements<K, V>* element = new Elements<K, V>(table_elements[i]->_key, table_elements[i]->_value);
			table[i] = element;
			delete table_elements[i];
		}
		else {
			table[i] = NULL;
		}
	}
	for (int i = size_table / 2; i < size_table; i++) {
		table[i] = NULL;
	}

	table_elements = new Elements<K, V>*[size_table];
	for (int i = 0; i < size_table; i++) {
		if (table[i] != NULL) {
			Elements<K, V>* element = new Elements<K, V>(table[i]->_key, table[i]->_value);
			table_elements[i] = element;
			table_elements[i]->flag = true;
			delete table[i];
		}
		else {
			table_elements[i] = NULL;
		}
	}

	delete[] table;
}


//метод добавления элемента - пары в таблицу
template <typename K, typename V>
void Hash_Map<K, V>::insert(K key, V value) {
	Elements<K, V>* element = new Elements<K, V>(key, value);
	size_t index = hash<K>()(key) % size_table;
	Elements<K, V>* cur_element = table_elements[index];
	bool collision_flag = true;
	if (cur_element == NULL) {
		//cout << "new element!!" << endl;

		table_elements[index] = element;
		table_elements[index]->flag = true;
		count_elements++;
		if (count_elements / size_table * 100 >= occupancy) {
			this->restructuring();
		}

	}

	else {
		//cout << "maybe collision..." << endl;
		if (collision_flag == 1) {

			for (int i = 0; i < size_table; i++) {
				int cur_index = (index + i) % size_table;
				Elements<K, V>* cur_element = table_elements[cur_index];
				if (cur_element != NULL) {
					if (cur_element->_key == key) {
						table_elements[cur_index]->_value = value;
						collision_flag = false;
					}
				}
			}
		}
		if (collision_flag != false) {
			//cout << "collision" << endl;
			this->collision(key, value, element);

		}
	}
}


//поиск уникального элемента
template <typename K, typename V>
void Hash_Map<K, V>::unique() {
	int count_unique = 0;
	for (int i = 0; i < size_table; i++) {
		if (table_elements[i] != NULL && table_elements[i]->flag != false) {
			int count_i = 0;
			for (int j = 0; j < size_table; j++) {
				if (i != j && table_elements[j] != NULL && table_elements[j]->flag != false) {
					if (table_elements[i]->_value == table_elements[j]->_value) {
						table_elements[j]->flag = false;
					}
				}
			}
			if (count_i == 0) {
				count_unique++;
			}
		}
	}

	cout << count_elements;
	cout << ' ';
	cout << count_unique << endl;
}


//MULTI HASH MAP
template <typename K, typename V>
class Multi_Hash_Map : public Hash_Map<K, V> {
public:
	Multi_Hash_Map(int size);
	void insert_multi(K key, V value);
	//V search_multi(K key);
	//void delete_multi(K key);
	Elements<K, V>** get_all_multi(K key) const;
	int sum_multi(K key) const;
};

template <typename K, typename V>
Multi_Hash_Map<K, V>::Multi_Hash_Map(int size) : Hash_Map<K, V>(size) {
};

//добавление элемента в multi hash map
template <typename K, typename V>
void Multi_Hash_Map<K, V>::insert_multi(K key, V value) {
	Elements<K, V>* element = new Elements<K, V>(key, value);
	size_t index = hash<K>()(key) % this->size_table;
	Elements<K, V>* cur_element = this->table_elements[index];
	if (cur_element == NULL) {
		//cout << "new element!!" << endl;
		this->table_elements[index] = element;
		this->table_elements[index]->flag = true;
		this->count_elements++;
		if (this->count_elements / this->size_table * 100 >= this->occupancy) {
			this->restructuring();
		}

	}
	else {
		this->collision(key, value, element);
	}
}


//количество элементов с заданным ключом
template <typename K, typename V>
int Multi_Hash_Map<K, V>::sum_multi(K key) const {
	size_t index = hash<K>()(key) % this->size_table;
	int sum = 0;
	for (int i = 0; i < this->size_table; i++) {
		int cur_index = (index + i) % this->size_table;
		Elements<K, V>* cur_element = this->table_elements[cur_index];
		if (cur_element != NULL && cur_element->flag != false) {
			if (cur_element->_key == key) {
				sum++;
			}
		}
	}
	return sum;
}


//получение всех элементов с заданным ключом
template <typename K, typename V>
Elements<K, V>** Multi_Hash_Map<K, V>::get_all_multi(K key) const {
	Elements<K, V>** elements = new Elements<K, V>*[sum_multi(key)];
	size_t index = hash<K>()(key) % this->size_table;
	int j = 0;
	for (int i = 0; i < this->size_table; i++) {
		int cur_index = (index + i) % this->size_table;
		Elements<K, V>* cur_element = this->table_elements[cur_index];
		if (cur_element != NULL && cur_element->flag != false) {
			if (cur_element->_key == key) {
				elements[j] = cur_element;
				j++;
			}
		}
	}
	return elements;
}


int main() {
	char type1, type2;
	cin >> type1 >> type2;
	int N;
	cin >> N;

	if (type1 == 'I' && type2 == 'I') {
		Hash_Map<int, int> hash_map(N);
		for (int i = 0; i < N; i++) {
			char action;
			cin >> action;
			if (action == 'A') {
				int k;
				int v;
				cin >> k >> v;
				hash_map.insert(k, v);
			}
			else if (action == 'R') {
				int k;
				cin >> k;
				hash_map.delete_element(k);
			}
		}
		//hash_map.print_table();
		hash_map.unique();
	}

	else if (type1 == 'I' && type2 == 'D') {
		Hash_Map<int, double> hash_map(N);
		for (int i = 0; i < N; i++) {
			char action;
			cin >> action;
			if (action == 'A') {
				int k;
				double v;
				cin >> k >> v;
				hash_map.insert(k, v);
			}
			else if (action == 'R') {
				int k;
				cin >> k;
				hash_map.delete_element(k);
			}
		}
		//hash_map.print_table();
		hash_map.unique();
	}

	else if (type1 == 'I' && type2 == 'S') {
		Hash_Map<int, string> hash_map(N);
		for (int i = 0; i < N; i++) {
			char action;
			cin >> action;
			if (action == 'A') {
				int k;
				string v;
				cin >> k >> v;
				hash_map.insert(k, v);
			}
			else if (action == 'R') {
				int k;
				cin >> k;
				hash_map.delete_element(k);
			}
		}
		//hash_map.print_table();
		hash_map.unique();
	}

	else if (type1 == 'D' && type2 == 'I') {
		Hash_Map<double, int> hash_map(N);
		for (int i = 0; i < N; i++) {
			char action;
			cin >> action;
			if (action == 'A') {
				double k;
				int v;
				cin >> k >> v;
				hash_map.insert(k, v);
			}
			else if (action == 'R') {
				double k;
				cin >> k;
				hash_map.delete_element(k);
			}
		}
		//hash_map.print_table();
		hash_map.unique();
	}

	else if (type1 == 'D' && type2 == 'D') {
		Hash_Map<double, double> hash_map(N);
		for (int i = 0; i < N; i++) {
			char action;
			cin >> action;
			if (action == 'A') {
				double k;
				double v;
				cin >> k >> v;
				hash_map.insert(k, v);
			}
			else if (action == 'R') {
				double k;
				cin >> k;
				hash_map.delete_element(k);
			}
		}
		//hash_map.print_table();
		hash_map.unique();
	}

	else if (type1 == 'D' && type2 == 'S') {
		Hash_Map<double, string> hash_map(N);
		for (int i = 0; i < N; i++) {
			char action;
			cin >> action;
			if (action == 'A') {
				double k;
				string v;
				cin >> k >> v;
				hash_map.insert(k, v);
			}
			else if (action == 'R') {
				double k;
				cin >> k;
				hash_map.delete_element(k);
			}
		}
		//hash_map.print_table();
		hash_map.unique();
	}

	else if (type1 == 'S' && type2 == 'I') {
		Hash_Map<string, int> hash_map(N);
		for (int i = 0; i < N; i++) {
			char action;
			cin >> action;
			if (action == 'A') {
				string k;
				int v;
				cin >> k >> v;
				hash_map.insert(k, v);
			}
			else if (action == 'R') {
				string k;
				cin >> k;
				hash_map.delete_element(k);
			}
		}
		//hash_map.print_table();
		hash_map.unique();
	}

	else if (type1 == 'S' && type2 == 'D') {
		Hash_Map<string, double> hash_map(N);
		for (int i = 0; i < N; i++) {
			char action;
			cin >> action;
			if (action == 'A') {
				string k;
				double v;
				cin >> k >> v;
				hash_map.insert(k, v);
			}
			else if (action == 'R') {
				string k;
				cin >> k;
				hash_map.delete_element(k);
			}
		}
		//hash_map.print_table();
		hash_map.unique();
	}

	else if (type1 == 'S' && type2 == 'S') {
		Hash_Map<string, string> hash_map(N);
		for (int i = 0; i < N; i++) {
			char action;
			cin >> action;
			if (action == 'A') {
				string k;
				string v;
				cin >> k >> v;
				hash_map.insert(k, v);
			}
			else if (action == 'R') {
				string k;
				cin >> k;
				hash_map.delete_element(k);
			}
		}
		//hash_map.print_table();
		hash_map.unique();
	}

	else if (type1 == 'M' && type2 == 'I') {
		Hash_Map<matrix, int> hash_map(N);
		for (int i = 0; i < N; i++) {
			char action;
			cin >> action;
			if (action == 'A') {
				matrix k;
				int v;
				int n;
				cin >> n;
				k.Input_matrix(n);
				cin >> v;
				hash_map.insert(k, v);
			}
			else if (action == 'R') {
				matrix k;
				int n;
				cin >> n;
				k.Input_matrix(n);
				hash_map.delete_element(k);
			}
		}
		//hash_map.print_table();
		hash_map.unique();
	}

	else if (type1 == 'M' && type2 == 'D') {
		Hash_Map<matrix, double> hash_map(N);
		for (int i = 0; i < N; i++) {
			char action;
			cin >> action;
			if (action == 'A') {
				matrix k;
				double v;
				int n;
				cin >> n;
				k.Input_matrix(n);
				cin >> v;
				hash_map.insert(k, v);
			}
			else if (action == 'R') {
				matrix k;
				int n;
				cin >> n;
				k.Input_matrix(n);
				hash_map.delete_element(k);
			}
		}
		//hash_map.print_table();
		hash_map.unique();
	}

	else if (type1 == 'M' && type2 == 'S') {
		Hash_Map<matrix, string> hash_map(N);
		for (int i = 0; i < N; i++) {
			char action;
			cin >> action;
			if (action == 'A') {
				matrix k;
				string v;
				int n;
				cin >> n;
				k.Input_matrix(n);
				cin >> v;
				hash_map.insert(k, v);
			}
			else if (action == 'R') {
				matrix k;
				int n;
				cin >> n;
				k.Input_matrix(n);
				hash_map.delete_element(k);
			}
		}
		//hash_map.print_table();
		hash_map.unique();
	}

	return 0;
}

/*
I S
5
A 783 Ivanov
A 999 Petrova
A 986 Sidorov
R 986
A 111 Kuznecov

M I
2
A 2 1 1 1 1 1
A 2 2 2 2 2 2



Multi_Hash_Map<int, string> a(10);
	string first = "first";
	string second = "second";
	string eps = "Epsilon";
	string again = "again";
	a.insert_multi(0, first);
	a.insert_multi(1, second);
	a.insert_multi(2, eps);
	a.insert_multi(2, again);
	a.print_table();
	string b = a.search(2);
	cout << b;
	//a.delete_element(2);
	cout << ' ' << endl;
	cout << ' ' << endl;
	//a.print_table();
	int c = a.sum_multi(2);
	cout << c;
	cout << endl;
	for (int i = 0; i < c; i++) {
		cout << a.get_all_multi(2)[i]->_value << endl;
	}

*/
```
