```python
#include <iostream>
#include <fstream>
#include <thread>
#include <mutex>
#include <chrono>
#include <string>
#include <ctime>
#include <set>

using namespace std;

int max_thread;
int count_thread = 0;
set<string>pages;
string From_test_data = "C:\\Users\\Ариничева Анна\\Desktop\\ConsoleApplication3\\test_data\\";
string To_new_data = "C:\\Users\\Ариничева Анна\\Desktop\\ConsoleApplication3\\new_data\\";


void crawler(string cur_page) {
	count_thread++;
	ifstream page;
	page.open(From_test_data + cur_page);
	string text;

	//считываем данные из файла
	getline(page, text);
	//while (!page.eof()) {
		//string line;
		//getline(page, line);
		//text += ('\n' + line);
	//}

	//копируем файл
	ofstream to;
	to.open(To_new_data + cur_page);
	to << text;
	to.close();

	//ищем номер первого символа первой ссылки в текущем файле
	size_t begin = text.find("<a href=\"file://", 0);

	//ищем ссылки на другие странички внутри текущей
	while (begin != string::npos) {
		size_t end = text.find("\">", begin);
		string href = text.substr(begin+16, end - begin-16);

		//если страница не была просмотрена ранее
		if ((pages).find(href) == pages.end()) {
			pages.insert(href);

			if (count_thread < max_thread) {
				//если число активных потоков меньше максимального числа потоков, то создаем новый поток и обрабатываем найденную ссылку там
				thread thread(crawler, href);
				thread.detach();
			}
			else {
				//иначе рекурсивно обрабатываем ссылку в текущем потоке
				crawler(href);
			}
		}

		begin = text.find("<a href=\"file://", begin += 5);
	}

	page.close();
}


int main() {
	int start_time = clock();
	string start_page; //стартовая страница
	cin >> start_page;
	cin >> max_thread; //количество рабочих потоков

	pages.insert(start_page);

	thread thread(crawler, start_page);
	thread.join();

	int end_time = clock();
	cout << pages.size() << endl;
	cout << (end_time - start_time) << endl;

	return 0;
}
```
